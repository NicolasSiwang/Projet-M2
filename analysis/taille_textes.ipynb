```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dossier = "SCOTU_data_txt_save/text"

fichiers = [f for f in os.listdir(dossier) if f.endswith('.txt')]

longueurs_textes = []

for fichier in fichiers:
    chemin_complet = os.path.join(dossier, fichier)
    with open(chemin_complet, 'r', encoding='utf-8') as f:
        contenu = f.read()
        longueurs_textes.append(len(contenu.split())) 

plt.figure(figsize=(10, 6))
sns.boxplot(x=longueurs_textes)

plt.title('Distribution des longueurs des textes dans les fichiers')
plt.xlabel('Longueur des textes (en mots)')
plt.ylabel('Fréquence')

plt.show()
```

```python
longueurs_textes = np.array(longueurs_textes)

stats = {
    "Nombre de fichiers": len(longueurs_textes),
    "Longueur minimale": np.min(longueurs_textes),
    "Longueur maximale": np.max(longueurs_textes),
    "Moyenne": np.mean(longueurs_textes),
    "Médiane": np.median(longueurs_textes),
    "Écart type": np.std(longueurs_textes),
    "1er Quartile (Q1)": np.percentile(longueurs_textes, 25),
    "3e Quartile (Q3)": np.percentile(longueurs_textes, 75)
}

print("Statistiques des longueurs des textes :")
for cle, valeur in stats.items():
    print(f"{cle}: {valeur}")
```
